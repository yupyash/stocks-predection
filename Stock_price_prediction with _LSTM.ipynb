```python
ma_day = [10, 20, 50] 

# Calculate moving averages and RSI for each company
for ma in ma_day:
    for company in company_list:
        column_name_ma = f"MA for {ma} days"
        company[column_name_ma] = company['Adj Close'].rolling(ma).mean()
```

```python

# Calculate RSI for each company
rsi_period = 14  # RSI period
for company in company_list:
    delta = company['Adj Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=rsi_period, min_periods=1).mean()
    avg_loss = loss.rolling(window=rsi_period, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    company['RSI'] = rsi
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from datetime import datetime
import yfinance as yf

# Define stock symbols
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'TSLA']

# Set up the start and end time for data fetch
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)

# Function to fetch stock data and prepare it
def prepare_data(stock_symbol, start, end, seq_length=60):
    df = yf.download(stock_symbol, start=start, end=end)
    data = df[['Adj Close']]  # Use adjusted close price for prediction
    
    # Normalize the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    
    # Prepare the sequences
    X, y = [], []
    for i in range(len(scaled_data) - seq_length):
        X.append(scaled_data[i:i + seq_length, 0])  # Last 60 days
        y.append(scaled_data[i + seq_length, 0])    # Next day's price
    X, y = np.array(X), np.array(y)
    
    # Reshape X to be 3D as LSTM expects (samples, timesteps, features)
    X = X.reshape(X.shape[0], X.shape[1], 1)
    
    return X, y, scaler, df

# Function to build and train LSTM model
def build_and_train_lstm(X_train, y_train, X_test, y_test):
    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(units=1))
    
    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
    
    return model

# Loop through each stock and train the model
for stock_symbol in tech_list:
    # Prepare the data for the current stock
    X, y, scaler, df = prepare_data(stock_symbol, start, end)
    
    # Split into train and test sets
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    # Build and train the LSTM model
    model = build_and_train_lstm(X_train, y_train, X_test, y_test)
    
    # Make predictions
    predicted_price = model.predict(X_test)
    
    # Inverse scale the predictions and actual values
    predicted_price = scaler.inverse_transform(predicted_price)
    y_test_scaled = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(y_test_scaled, color='blue', label='Actual Stock Price')
    plt.plot(predicted_price, color='red', label='Predicted Stock Price')
    plt.title(f'{stock_symbol} Stock Price Prediction (LSTM)')
    plt.xlabel('Time')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.show()
```

```python
# Create the training data set 
# Create the scaled training data set
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()
        
# Convert the x_train and y_train to numpy arrays 
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
# x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1), activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(25, activation='relu'))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=5, epochs=25)  # Adjust epochs and batch size as needed
```

```python
# Get the model's predicted price values (only 'Close' prices)
predictions = model.predict(x_test)

# Reshape predictions to match the scaler's expected input
predictions = np.reshape(predictions, (predictions.shape[0], 1))

# Inverse transform the predictions to the original scale (matching the number of features)
# Adding zeros for non-close features
predictions = scaler.inverse_transform(np.hstack((predictions, np.zeros((predictions.shape[0], 4)))))

# Since y_test only contains 'Close' values, we need to inverse transform y_test as well
y_test_scaled = scaler.inverse_transform(np.hstack((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 4)))))

# Calculate RMSE using only the 'Close' values (first column of predictions and y_test_scaled)
rmse = np.sqrt(np.mean(((predictions[:, 0] - y_test_scaled[:, 0]) ** 2)))
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
